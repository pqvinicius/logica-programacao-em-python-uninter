##**Ordenação e ADD elementos**

```python
rng = np.random.default_rng()
array_1 = rng.random([2,5])
print(f'Array 1 SEM ORDENAÇÃO:\n{array_1}\n')
print(f'Array 1 COM ORDENAÇÃO:\n{np.sort(array_1)}\n')
print(f'Array 1 COM ORDENAÇÃO:\n{np.sort(array_1,  axis = 0)}\n')
print(f'Array 1 COM ORDENAÇÃO:\n{np.sort(array_1, axis = 1)}\n')

# CONCATENAÇÃO

array_2 = rng.random([2,1])
print(f'ARRAY 2:\n{array_2}\n')

array_concatenado = np.concatenate([array_1,array_2], axis = 1)
print(f'Array Concatenado:\n{array_concatenado}\n')

array_concatenado_2 = np.hstack([array_1,array_2])
print(f'Array Concatenado:\n{array_concatenado_2}\n')
```

##**Como mudar o formato de um array**

```python
x = np.arange(0,20)
print(f'Array x:\n{x}\n')
x = x.reshape(4,5)
print(f'Array x:\n{x}\n')
x = x.flatten()
print(f'Array x:\n{x}\n')
y=x.reshape([4,5])
print(f'Array y:\n{y}\n')
y=np.vsplit(y,4)
print(f'Array y:\n{y}\n')
```

## PLOTLY

```python
#Criando arrays para montar

a = 3/4
b = 2/4
c = 1/4
vetor_a=np.ones(20)
vetor_b=np.ones(20)
vetor_c=np.ones(20)

for i in range (20):
  vetor_a[i]= vetor_a[i - 1] = a**i
  vetor_b[i]= vetor_b[i - 1] = b**i
  vetor_c[i]= vetor_c[i - 1] = c**i
# print(vetor_a)
# print(vetor_b)
# print(vetor_c)

np.savetxt('vetor_a.txt', vetor_a, fmt ='%f', delimiter=';')
np.savetxt('vetor_b.txt', vetor_b, fmt ='%f', delimiter=';')
np.savetxt('vetor_c.txt', vetor_c, fmt ='%f', delimiter=';')
```

##Plotly - Criando nosso primeiro grafico

```python
array_a = np.loadtxt('vetor_a.txt', dtype = np.float64, delimiter = ';')
array_b = np.loadtxt('vetor_b.txt', dtype = np.float64, delimiter = ';')
array_c = np.loadtxt('vetor_c.txt', dtype = np.float64, delimiter = ';')

# print(array_a)
# print(array_b)
# print(array_c)

array_abc = np.vstack([array_a, array_b, array_c])
print(array_abc)
array_abc = array_abc.transpose()
fig = px.scatter(array_abc,
                 title = 'Gráfico de dispersão teste')
fig.show()
fig2 = px.line(array_abc,
               title = 'Gráfico de linha teste')
fig2.show()
```
