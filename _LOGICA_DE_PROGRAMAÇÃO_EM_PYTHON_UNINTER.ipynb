
# **AULA 1** - trabalhando com numpy e plotly

```python
#importações de libs
import numpy as np
import plotly
```

```python
#diferença entre array numpy para list in python

#lista normal
num_elementos = 10000000
lista_elementos =[]
for x in range(num_elementos):
  lista_elementos.append(x)
print(lista_elementos[-1])
```

```python
# lista numpy
array_elementos = np.arange(num_elementos)
print(array_elementos[-1])
```

## **Criação de Arrays em Numpy**

```python

array_teste = np.array([1,2,3,4,5,7,8,9,10])
print(f'COM VARIAVEL: Criação de array numpy utilizando o método .array:\n{array_teste}\n')
print(f'DIRETO NO PRINT: Criação de array numpy utilizando o método .array:\n{np.array([1,2,3,4,5,7,8,9,10])}\n')
print(f'DIRETO NO PRINT: Criação de array numpy de zeros utilizando o método .zeros:\n{np.zeros([10])}\n')
print(f'DIRETO NO PRINT: Criação de array numpy de uns utilizando o método .ones:\n{np.ones([10])}\n')
print(f'DIRETO NO PRINT: Criação de array numpy de uns utilizando o método .empy:\n{np.empty([10])}\n')
print(f'DIRETO NO PRINT: Criação de array numpy de uns utilizando o método .arange:\n{np.arange(10)}\n')
print(f'DIRETO NO PRINT: Criação de array numpy de uns utilizando o método .linspace:\n{np.linspace(0,100,num=101)}\n')

```
